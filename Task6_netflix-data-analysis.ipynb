```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
data=pd.read_csv('netflix.csv')
```

```python
data.columns
```

```python
data.head()
```

```python
data.isnull().sum().sum()
```

```python
data.info()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

```python
data['premiere']=pd.to_datetime(data['premiere'])
```

```python
data['premiere_month']=data['premiere'].dt.month
data['premiere_day']=data['premiere'].dt.day
data['premiere_year']=data['premiere'].dt.year
```

```python
data.head()
```

```python
data.tail()
```

```python
data.describe()
```

```python
data.hist(bins=50,grid=True,color='chocolate',figsize=(15,9))
plt.show()
```

```python
data.head(8)
```

```python
data.language.value_counts().head(10).plot(kind = 'pie',colors=sns.color_palette('viridis'))
plt.title("Types of Languages in movies")
plt.tight_layout() 
```

```python
plt.figure(figsize=(10,5))
data.genre.value_counts().head(15).plot(kind = 'bar',color=sns.color_palette('husl'))
plt.title("Types of Genres of movies")
plt.xticks(rotation=45)
plt.tight_layout() 
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x=data['premiere_year'],data=data, order=data['premiere_year'].value_counts().index,palette='mako')
plt.xticks(rotation=45)
plt.title('Yearly premiere Count')
plt.show()
```

```python
plt.figure(figsize=(12,4))
sns.barplot(data=data,x='year',y='runtime',palette='magma')
plt.title('Yearly Run Time')
plt.show()
```

```python
plt.figure(figsize=(9,4))
sns.lineplot(data=data,x='imdb_score',y='runtime',color='olive')
plt.title('Score on the basis of Run Time')
plt.show()
```

```python
plt.figure(figsize=(12,4))
top_6_movies = data.sort_values(by='imdb_score', ascending=False).head(6)
sns.barplot(x=top_6_movies['title'],y=top_6_movies['imdb_score'],palette='icefire')
plt.xlabel('Movie Name')
plt.ylabel('Score')
plt.title('Top 6 Movies')
plt.xticks(rotation=25)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(17,5))
sns.barplot(x=data['language'],y=data['imdb_score'],palette='rocket',hue=data['language'])
plt.xlabel('Language Name')
plt.ylabel('Score')
plt.title('Score on the basis of Language')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
