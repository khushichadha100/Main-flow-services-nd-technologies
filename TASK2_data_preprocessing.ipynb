```python
file=file.drop_duplicates()  #dropping the duplicate columns to  clean the data more effeciently
file
```

```python
file2=file.fillna(value=0) #filling all the null values by 0
file2
```

```python
file2.isnull().sum().sum() #after filling all null values by 0 , theres no null data left
```

```python
file3=file.drop(["Observation"],axis=1) #deleting the date time data type column
file3=file3.fillna(file3.mean()) #filling the nan values with the mean of their column
file3
```

```python
file3.isnull().sum().sum() # no data fields are  null now hence its the perfect file for further analysis.
```

```python
file4=file.fillna(method="bfill") #using backward filling to fill null data
file4
```

```python
file4.isnull().sum().sum() #still 2 values are null
```

```python
file5=file.fillna(method="ffill") #using forward filling to fill null data
file5
```

```python
file5.isnull().sum().sum() #still 2 values are null
```

```python
file3.columns #extracting all the columns from file 3 because it contains no null values as now
```

```python
file3.describe() #need decription  of the data for quantile analysis to detect outlier
```

```python
Q1=file3.quantile(0.25)
Q3=file3.quantile(0.75)
IQR=Q3-Q1
IQR      #detecting outlier using IQR
```

```python
file3=file3[~((file3<(Q1-1.5*IQR))|(file3>(Q3+1.5*IQR))).any(axis=1)]    #checking through formula
file3
```
